```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
files = ['plastic_out.csv', 'plastic_lagged_out.csv', 'plastic_npr_out.csv']
dbs = []
for file in files:
  dbs.append((pd.read_csv(file), file))
```

```python
plt.rcParams.update({'font.size': 18})
fig = plt.figure(figsize = (10, 7))
ax = plt.axes()
for db, name in dbs:
  ax.plot(db['time'], db['max_ep'], label=name)
```
