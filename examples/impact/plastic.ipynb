```python
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob
```

```python
files = ['plastic_out.csv', 'plastic_lagged_out.csv', 'plastic_lagged_stabilized_out.csv', 'plastic_npr_out.csv', 'plastic_npr_stabilized_out.csv', 'plastic_npr_stabilized_2_out.csv', 'plastic_npr_stabilized_0.5_out.csv', 'plastic_npr_stabilized_3_out.csv']
dbs = []
for file in files:
  dbs.append((pd.read_csv(file), file))
```

```python
plt.rcParams.update({'font.size': 12})
fig = plt.figure(figsize = (10, 7))
ax = plt.axes()
for name in glob('plastic_npr_stabilized_*_out.csv'):
  db = pd.read_csv(name)
  ax.plot(db['time'], db['rear_x'], label=name)
ax.legend()
```

```python
plt.rcParams.update({'font.size': 12})
fig = plt.figure(figsize = (10, 7))
ax = plt.axes()
for db, name in dbs:
  ax.plot(db['time'], db['vel_x'], label=name)
ax.legend()
```
